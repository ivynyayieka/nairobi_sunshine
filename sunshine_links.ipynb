```python
df['city_url']=df.sun_city.str.replace(' ', '-').str.lower()


df['country_url']=df.sun_country.str.replace(' ', '-').str.lower()
```

```python
example_url="https://www.sunrise-and-sunset.com/en/sun/kenya/nairobi/2023/december"

df["dec_url_city"]="https://www.sunrise-and-sunset.com/en/sun/"+df.country_url+"/"+df.city_url+"/2023/december"

df["dec_url_city"]=df["dec_url_city"].str.replace('\n', '')

df
```

```python
def get_dec_june_raw_html(x):
    try:
        raw_html=requests.get(x).content
        return raw_html
    except:
        pass
    
df ['dec_raw_html'] = df.dec_url_city.apply(get_dec_june_raw_html)
```

```python
#for NAirobi
BeautifulSoup(df.dec_raw_html[0], "html.parser").find_all('tr')[22].find_all('td')[3].text.strip()
```

```python
#for dec soup_doc
def get_dec_june_beautiful_soup(x):
    try:
        soup_doc=BeautifulSoup(x, "html.parser")
        return soup_doc
    except:
        pass
    
df ['dec_soup_doc'] = df.dec_raw_html.apply(get_dec_june_beautiful_soup)
```

```python
#for dec_duration by scraping

def daylight_june_dec(x):
    try:
        duration= x.find_all('tr')[22].find_all('td')[3].text.strip()
        return duration
    except:
        pass

df ['dec_duration'] = df.dec_soup_doc.apply(daylight_june_dec)

df ['dec_duration']
df

#how do you try except within a lambda function 

#move it out of lambda?


```
